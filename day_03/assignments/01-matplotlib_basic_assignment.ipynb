## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
# Write your code here
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('../data/company_sales_data.csv')
month_number = dataset['month_number']
total_profit = dataset['total_profit']
plt.plot(month_number, total_profit)
plt.xlabel('Month Number',size=12)
plt.ylabel('Profit in dollar',size=12)
plt.title('Company profit per month', size=14)
plt.ylim(100000,500000)
```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

**Expected output**

<img src="../data/company_sales_data_of_last_year.png" alt="company_sales_data_of_last_year" style="width: 700px;"/>


```python
# write your code here
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('../data/company_sales_data.csv')
month_number = dataset['month_number']
total_units = dataset['total_units']
plt.figure(figsize=(10, 4))
plt.plot(month_number, total_units,linewidth=3,color='red', linestyle='dashed', marker='o',
     markerfacecolor='black', markersize=5,label = " Profit data of last year")
plt.xlabel('Month Number',size=12)
plt.ylabel('Profit in dollar',size=12)
plt.title('Company sales data of last year', size=14)
plt.legend(loc="lower right")
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
# write your code here
# Import relevant modules
import pandas as pd
import matplotlib.pyplot as plt

# Import dataset
data = pd.read_csv('../data/company_sales_data.csv')
month_number= data['month_number']
facecream = data['facecream']
facewash = data['facewash']
toothpaste = data['toothpaste']
bathingsoap = data['bathingsoap']
shampoo = data['shampoo']
plt.figure(figsize=(7, 4))
# Plot the data series
plt.plot(month_number, facecream,marker='o',markersize=3, label='facecream sales data')
plt.plot(month_number, facewash,marker='o',markersize=3, label='facewash sales data')
plt.plot(month_number, toothpaste,marker='o',markersize=3, label='toothpaste sales data')
plt.plot(month_number, bathingsoap, marker='o',markersize=3,label='bathingsoap sales data')
plt.plot(month_number, shampoo,marker='o',markersize=3, label='shampoo sales data')
# Label the x- and y-axes
plt.xlabel('Month number',size=12)
plt.ylabel('Sales units in number',size=12)


# Add the title and legend
plt.title('Sales data', size=14)
plt.legend( prop={'size': 7})

# Show the figure
plt.show()
```
