## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
# write your code here
# Import relevant modules
import pandas as pd
import matplotlib.pyplot as plt

# Import dataset
data = pd.read_csv('../data/company_sales_data.csv')
month_number= data['month_number']
facecream = data['facecream']
facewash = data['facewash']
toothpaste = data['toothpaste']
bathingsoap = data['bathingsoap']
shampoo = data['shampoo']
plt.figure(figsize=(7, 4))
# Plot the data series
plt.plot(month_number, facecream,marker='o',markersize=3, label='facecream sales data')
plt.plot(month_number, facewash,marker='o',markersize=3, label='facewash sales data')
plt.plot(month_number, toothpaste,marker='o',markersize=3, label='toothpaste sales data')
plt.plot(month_number, bathingsoap, marker='o',markersize=3,label='bathingsoap sales data')
plt.plot(month_number, shampoo,marker='o',markersize=3, label='shampoo sales data')
# Label the x- and y-axes
plt.xlabel('Month number',size=12)
plt.ylabel('Sales units in number',size=12)


# Add the title and legend
plt.title('Sales data', size=14)
plt.legend( prop={'size': 7})

# Show the figure
plt.show()
```

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
# write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Import dataset
dataset = pd.read_csv('../data/company_sales_data.csv')
month_number= dataset['month_number']
facecream = dataset['facecream']
facewash = dataset['facewash']
plt.figure(figsize=(9, 4))
width = 0.3
plt.bar(month_number - width/2,facecream,width=width, label = 'Facecream sales data')
plt.bar(month_number + width/2,facewash,width=width, label = 'Facewash sales data')

plt.xlabel('Month number',size=12)
plt.ylabel('Sales units in number',size=12)
plt.title('Facecream and facewash sales data', size=14)
plt.xticks(month_number)
plt.grid(linestyle = '--')
plt.legend(loc = "upper left")
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
# write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Import dataset
data = pd.read_csv('../data/company_sales_data.csv')
month_number= data['month_number']
facecream = data['facecream']
facewash = data['facewash']
toothpaste = data['toothpaste']
bathingsoap = data['bathingsoap']
shampoo = data['shampoo']
moisturizer = data['moisturizer']

plt.figure(figsize=(10, 5))
plt.plot([], [],color="purple", label='facecream')
plt.plot([], [],color="blue", label='facewash')
plt.plot([], [],color="red", label='toothpaste')
plt.plot([], [],color="black", label='bathingsoap')
plt.plot([], [],color="green", label='shampoo')
plt.plot([], [],color="yellow", label='moisturizer')

plt.yticks([5000,10000,15000,20000,25000,30000])
plt.stackplot(month_number,facecream,facewash,toothpaste,bathingsoap,shampoo,moisturizer,colors=['purple','blue','red','black','green','yellow'])
# Label the x- and y-axes
plt.xlabel('Month number',size=12)
plt.ylabel('Sales units in number',size=12)


# Add the title and legend
plt.title('All product sales data using stack plot', size=14)
plt.legend( loc = "upper left", prop = {'size': 7} )

# Show the figure
plt.show()
```
